9. Assign to a variable g the expression e^x-sin(x). Evaluate this expression in x = 0. Compute its first order derivative and then evaluate it in x = 0. Find a primitive. Evaluate integral from -1 to 1 of gdx. Assign to a variable f the second order derivative of g and evaluate it in x = 0.

```sagemath
var('x')
g(x) = exp(x) - sin(x)
g(x).subs(x=0)
diff(g(x), x, 2)
integrate(g(x), x)
integrate(g(x), x, -1, 1)
f(x) = diff(g(x), x, 3)
f(x).subs(x=0)
```

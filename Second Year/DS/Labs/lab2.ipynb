6. x" \+ x = 0, x\(π/2\) = 1, x'\(π/2\) = −2. Show that x\(t\) = √5\*cos\(t − arctan1/2\) \(In Maple you can use ”expand”, in Sage ”trig simplify”\). What is the amplitude of the oscillation?



```python
var('t')
x = function('x')(t)
eq6 = diff(x, t, 2) + diff(x, t) == 0
sol = desolve(eq6, x, ics=[pi/2, 1, -2]) # ics = [t0, x(t0), x'(t0)]
(sqrt(5)*cos(t - arctan(1/2))).trig_simplify()
plot(sol, -10, 10)
```
